```python
def downcast(df):
    float_cols = [c for c in df if df[c].dtype in ["float64", "float32"]]
    int_cols = [c for c in df if df[c].dtype in ["int64", "int32"]]
    df[float_cols] = df[float_cols].astype(np.float16)
    df[int_cols] = df[int_cols].astype(np.int16)
    types = {}
    if "month" in df.columns:
        if df.month.isnull().any() == False:
            types["month"] = np.uint8
    if "year" in df.columns:
        if df.year.isnull().any() == False:

            types["year"] =  np.uint8
    if "date_block_num" in df.columns:
        types["date_block_num"] = np.int8
    if "shop_id" in df.columns:
        types["shop_id"] = np.int8
    if "item_id" in df.columns:
        types["item_id"] = np.int16
    if "item_cnt_month" in df.columns:
        ##clipped between 0 to 20
        types["item_cnt_month"] = np.int8
    df= df.astype(types)

    return df
# dataDf.item_cnt_month = dataDf.item_cnt_month.fillna(0)
# dataDf= dataDf.astype({'month': np.uint8, 'year': np.uint16, 'item_id': np.uint16, 'date_block_num': np.uint8, "shop_id": np.uint8})
# dataDf = downcast(dataDf)
# dataDf.item_cnt_month = dataDf.item_cnt_month.map(int).astype(np.int16)
```

```python
import mlflow
import time
import logging
import numpy as np
import xgboost as xgb
from xgboost import XGBRegressor, DMatrix
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
from sklearn.metrics import mean_squared_error
from math import sqrt
import json

config = json.load(open("config.json", "rb"))
mlflow_url = config["MLFLOW_URL"]
mlflow.set_tracking_uri(mlflow_url)


logging.basicConfig(filename='mlflow.log',level=logging.DEBUG)

ts = time.time()
space = {
    'num_boost_round': hp.choice('num_boost_round', np.arange(500, 1001, dtype=int)),
    'eta': hp.quniform('eta', 0.025, 0.5, 0.025),
    'max_depth':  hp.choice('max_depth', np.arange(6, 14, dtype=int)),
    'min_child_weight': hp.quniform('min_child_weight', 1, 6, 1),
    'subsample': hp.quniform('subsample', 0.5, 1, 0.05),
    'gamma': hp.quniform('gamma', 0.5, 1, 0.05),
    'colsample_bytree': hp.quniform('colsample_bytree', 0.5, 1, 0.05),

}


def score(params):
    try:
        print (params)
        mlflow.set_experiment("Predict Future Sales 3")
        with mlflow.start_run() as run:
            mlflow.log_params(params)
            
            dtrain = DMatrix(trainX, label=trainY)
            dvalid = DMatrix(valX, label=valY)
#             model = XGBRegressor(
# #                 **params,
#                 seed=42)
            logging.info("Start training...")
            model = xgb.train(params, dtrain, params["num_boost_round"], evals=[(dtrain, 'train'), (dvalid, 'eval')], verbose_eval = True, early_stopping_rounds = 10)
#             model.fit(
#                 trainX, 
#                 trainY, 
#                 eval_metric="rmse", 
#                 eval_set=[(trainX, trainY), (valX, valY)], 
#                 verbose=True, 
#                 early_stopping_rounds = 10, )
            logging.info("Finish training")
            logging.info("Calculate score...")
            predictions = model.predict(dvalid, ntree_limit=model.best_iteration)
            score = sqrt(mean_squared_error(valY.values, predictions))
            logging.info("Finished calculating")
            loss = 1 - score
            mlflow.log_metrics({
                "val_loss": loss,
                "val_rmse": score,
                "best_iteration": model.best_iteration
            })
            return {'loss': loss, 'status': STATUS_OK}
    except Exception as e:
        logging.warning(e)

        

trials = Trials()

best = fmin(score, space, algo=tpe.suggest, 
            trials=trials, 
            max_evals=250)

time.time() - ts
```
