```python
def downcast(df):
    float_cols = [c for c in df if df[c].dtype in ["float64", "float32"]]
    int_cols = [c for c in df if df[c].dtype in ["int64", "int32"]]
    df[float_cols] = df[float_cols].astype(np.float16)
    df[int_cols] = df[int_cols].astype(np.int16)
    types = {}
    if "month" in df.columns:
        if df.month.isnull().any() == False:
            types["month"] = np.uint8
    if "year" in df.columns:
        if df.year.isnull().any() == False:

            types["year"] =  np.uint8
    if "date_block_num" in df.columns:
        types["date_block_num"] = np.int8
    if "shop_id" in df.columns:
        types["shop_id"] = np.int8
    if "item_id" in df.columns:
        types["item_id"] = np.int16
    if "item_cnt_month" in df.columns:
        ##clipped between 0 to 20
        types["item_cnt_month"] = np.int8
    df= df.astype(types)

    return df
# dataDf.item_cnt_month = dataDf.item_cnt_month.fillna(0)
# dataDf= dataDf.astype({'month': np.uint8, 'year': np.uint16, 'item_id': np.uint16, 'date_block_num': np.uint8, "shop_id": np.uint8})
# dataDf = downcast(dataDf)
# dataDf.item_cnt_month = dataDf.item_cnt_month.map(int).astype(np.int16)
```

```python
trainDf = dataDf[dataDf.date_block_num < 33]
trainDf = trainDf[trainDf.date_block_num > 11]
valDf = dataDf[dataDf.date_block_num == 33]
testDf = dataDf[dataDf.date_block_num == 34]
trainDf.info()
```

```python
trainX = trainDf.drop(["item_cnt_month"], axis = 1)
trainY = trainDf.item_cnt_month
trainX
```

```python
valX = valDf.drop(["item_cnt_month"], axis = 1)
valY = valDf.item_cnt_month
valX
```

```python
%%time
params = {
    'num_boost_round': 23,
    'eta': 0.1,
    'max_depth': 13,
    'min_child_weight': 4,
    'subsample':0.8,
    'gamma': 0.7,
    'colsample_bytree': 0.55

}
dtrain = DMatrix(trainX, label=trainY)
dvalid = DMatrix(valX, label=valY)
model = xgb.train(params, dtrain, params["num_boost_round"], evals=[(dtrain, 'train'), (dvalid, 'eval')], verbose_eval = True, early_stopping_rounds = 10)
# model = xgb.train(params, dtrain, params["num_boost_round"], evals=[(dtrain, 'train')], verbose_eval = True, early_stopping_rounds = 10)
```
