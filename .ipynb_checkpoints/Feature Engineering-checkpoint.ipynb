```python
def downcast(df):
    float_cols = [c for c in df if df[c].dtype in ["float64", "float32"]]
    int_cols = [c for c in df if df[c].dtype in ["int64", "int32"]]
    df[float_cols] = df[float_cols].astype(np.float16)
    df[int_cols] = df[int_cols].astype(np.int16)
    types = {}
    if "month" in df.columns:
        if df.month.isnull().any() == False:
            types["month"] = np.uint8
    if "year" in df.columns:
        if df.year.isnull().any() == False:

            types["year"] =  np.uint8
    if "date_block_num" in df.columns:
        types["date_block_num"] = np.int8
    if "shop_id" in df.columns:
        types["shop_id"] = np.int8
    if "item_id" in df.columns:
        types["item_id"] = np.int16
    if "item_cnt_month" in df.columns:
        ##clipped between 0 to 20
        types["item_cnt_month"] = np.int8
    df= df.astype(types)

    return df
# dataDf.item_cnt_month = dataDf.item_cnt_month.fillna(0)
# dataDf= dataDf.astype({'month': np.uint8, 'year': np.uint16, 'item_id': np.uint16, 'date_block_num': np.uint8, "shop_id": np.uint8})
# dataDf = downcast(dataDf)
# dataDf.item_cnt_month = dataDf.item_cnt_month.map(int).astype(np.int16)
```

```python
import statistics as stats

def rolling_mean(df, merge_on_columns = ["shop_id", "item_id"], roll_col = "item_cnt_month", col_name = str, period = int):
    rollDf = df[["date_block_num", *merge_on_columns, roll_col]]
    rollDf.groupby([*merge_on_columns, "date_block_num"]).sum()
    rollDf = rollDf.groupby([*merge_on_columns, "date_block_num"]).sum().rolling(period).mean().reset_index()
    rollDf[roll_col] = rollDf[roll_col].shift(1)
    rollDf.rename(columns = {roll_col: f"{col_name}_ma_{period}"}, inplace = True)    
    df = df.merge(rollDf, on=["date_block_num", *merge_on_columns], how="left", suffixes = ("", ""))
    return df

def rolling_var(df, merge_on_columns = ["shop_id", "item_id"], roll_col = "item_cnt_month", col_name = str, period = int):
    rollDf = df[["date_block_num", *merge_on_columns, roll_col]]
    rollDf.groupby([*merge_on_columns, "date_block_num"]).sum()
    rollDf = rollDf.groupby([*merge_on_columns, "date_block_num"]).sum().rolling(period).std().reset_index()
    rollDf[roll_col] = rollDf[roll_col].shift(1)
    rollDf.rename(columns = {roll_col: f"{col_name}_var_{period}"}, inplace = True)
    df = df.merge(rollDf, on=["date_block_num", *merge_on_columns], how="left", suffixes = ("", ""))
    return df

def rolling_median(df, merge_on_columns = ["shop_id", "item_id"], roll_col = "item_cnt_month", col_name = str, period = int):
    rollDf = df[["date_block_num", *merge_on_columns, roll_col]]
    rollDf = rollDf.groupby([*merge_on_columns, "date_block_num"]).sum().rolling(period).median().reset_index()
    rollDf[roll_col] = rollDf[roll_col].shift(1)
    rollDf.rename(columns = {roll_col: f"{col_name}_median_{period}"}, inplace = True)
    df = df.merge(rollDf, on=["date_block_num", *merge_on_columns], how="left", suffixes = ("", ""))
    return df
```

```python
%%time
import time

def shop_last_sale(dataDf):
#     dataDf.at[7323371, "item_cnt_month"] = 1
#     global cache
    cache = {}
    dataDf['item_shop_last_sale'] = -1
    dataDf['item_shop_last_sale'] = dataDf['item_shop_last_sale'].astype(np.int8)
    for idx, row in dataDf.iterrows():    
        key = str(row.item_id)+' '+str(row.shop_id)

        if key in cache:
            last_date_block_num = cache[key]
            dataDf.at[idx, 'item_shop_last_sale'] = row.date_block_num - last_date_block_num
#                 cache[key] = row.date_block_num     
        if row.item_cnt_month!=0:
            cache[key] = row.date_block_num
    return dataDf
dataDf = shop_last_sale(dataDf)
dataDf = downcast(dataDf)
```

```python
%%time
def last_sale(dataDf):
#     dataDf.at[7323371, "item_cnt_month"] = 1

    cache = {}
    dataDf['item_last_sale'] = -1
    dataDf['item_last_sale'] = dataDf['item_last_sale'].astype(np.int8)
    for idx, row in dataDf.iterrows():    
        key = row.item_id
        if key not in cache:
            if row.item_cnt_month!=0:
                cache[key] = row.date_block_num
        else:
            last_date_block_num = cache[key]
            if row.date_block_num>last_date_block_num:
                dataDf.at[idx, 'item_last_sale'] = row.date_block_num - last_date_block_num
                cache[key] = row.date_block_num         
    return dataDf
dataDf = last_sale(dataDf)
dataDf = downcast(dataDf)
```

```python
%%time
# def shift_and_pct_change(df, merge_on_columns = ["shop_id", "item_id"], merge_column = str, column_name = str, lags = list):
#     '''
#     Add the previous monthly sales value for that lag period, and also the month pct_change of sales 
#     '''
#     ori_length = len(df)
#     for lag in lags:
#         df = lag_shift(df, merge_on_columns = merge_on_columns, merge_column = merge_column, column_name = f"{column_name}_{lag}", lag = lag)
#         df = lag_shift(df, merge_on_columns= merge_on_columns, merge_column = merge_column, column_name = f"{column_name}_{lag+1}", lag = lag + 1)
#         df.columns
#         df[f"delta_{column_name}_{lag}"] = (df[f"{column_name}_{lag}"]/ df[f"{column_name}_{lag + 1}"]) - 1
#         df[f"delta_{column_name}_{lag}"] = df[f"delta_{column_name}_{lag}"].astype(np.float32)
#         df.drop([f"{column_name}_{lag + 1}"], axis = 1, inplace = True)
# #         print (df.head(1))

#     assert len(df) == ori_length, f"{len(df)} {ori_length}"
#     return df

def lag_shift(df, merge_on_columns = ["shop_id", "item_id"], merge_column = str, column_name = str, lag = int):
    '''
    column: The column to shift
    lag: Number of month to shift
    '''
    tmpDf = df[["date_block_num", *merge_on_columns, merge_column]].drop_duplicates()
    tmpDf["date_block_num"] = tmpDf["date_block_num"].apply(lambda x: x+ lag)
    df = df.merge(tmpDf, on=["date_block_num", *merge_on_columns], how="left", suffixes= ("", f"_{lag}"))
    df.rename(columns = {f"{merge_column}_{lag}": column_name}, inplace = True)
    df[column_name] = df[column_name].fillna(0)
#     df = downcast(df)
    return df
def shift_and_abs_change(df, merge_on_columns = ["shop_id", "item_id"], merge_column = str, column_name = str, lags = list, dtype = np.float16):
    '''
    Add the previous monthly sales value for that lag period, and also the month pct_change of sales 
    '''
    ori_length = len(df)
    for lag in lags:
        df = lag_shift(df, merge_on_columns = merge_on_columns, merge_column = merge_column, column_name = f"{column_name}_{lag}", lag = lag)
        df = lag_shift(df, merge_on_columns= merge_on_columns, merge_column = merge_column, column_name = f"{column_name}_{lag+1}", lag = lag + 1)
        df[f"{column_name}_{lag}"] = df[f"{column_name}_{lag}"].astype(dtype)
        df[f"delta_{column_name}_{lag}"] = df[f"{column_name}_{lag}"] - df[f"{column_name}_{lag + 1}"]
        df[f"delta_{column_name}_{lag}"] = df[f"delta_{column_name}_{lag}"].astype(dtype)
        df.drop([f"{column_name}_{lag + 1}"], axis = 1, inplace = True)
#         print (df.head(1))
    assert len(df) == ori_length, f"{len(df)} {ori_length}"
    return df

params = {
    "df": dataDf,
    "merge_on_columns":["shop_id", "item_id"],
    "merge_column": "item_cnt_month",
    "column_name": "shop_item_cnt_month",
    "lags": [1,2] ,
    "dtype": np.int8
}
dataDf = shift_and_abs_change(**params)
dataDf = downcast(dataDf)
dataDf.info()
```
